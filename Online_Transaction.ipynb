```python
# Import libraries 
import matplotlib.pyplot as plt #used for plotting data 
import pandas as pd
import numpy as np
from datetime import datetime
mydata= pd.read_csv("C:\\Users\\HP\\Desktop\\Transaction.csv")  ##Importing the Dataset from my local machine
mydata
```

```python
mydata.isnull().sum() ###Checking for Null values
```

```python
###Data Cleaning
mydata.dropna()   #Removing the null values
```

```python
##Calculating the revenue for each sale

mydata['Revenue'] = mydata['Price'] * mydata['Quantity']
mydata.head(20)
```

```python
mydata['month'] = pd.DatetimeIndex(mydata['Date']).month
mydata
```

```python
##Grouping the Transacions
mydata.groupby([ 'CustomerNo','TransactionNo', 'ProductName']).sum()
```

```python
Product_Sale = mydata.groupby('ProductName').agg(Quant = ('Quantity', 'sum'))
Product_Sale.sort_values(by = 'Quant', ascending = False).head(10).plot(kind= 'bar',stacked=True, figsize=(10,8),color='green', width=0.5)
plt.xlabel('Product Name')
plt.ylabel('Quantiy Sold')
plt.title('Bar Chart Showing Most Sold Products')
```

```python
##Sales per month
#converting date to only year and month
mydata['Date'] = pd.to_datetime(mydata['Date'])
mydata['Year_Month'] = mydata['Date'].dt.to_period('M')
mydata
```

```python
sales_per_month = mydata.groupby('Year_Month').agg(num_sales_month = ('Revenue', 'sum')).plot(kind= 'line', stacked=True, figsize=(10,8),color='green')
plt.xlabel('Month and Year')
plt.ylabel('Number of Sales')
plt.title('Line Graph Showing Sales Trend Over the Months')
sales_per_month 
```

```python
customer_purchase = mydata.groupby('TransactionNo').agg(Quant = ('Quantity', 'sum'))
customer_purchase.sort_values(by = 'Quant', ascending = False).head(10)
```

```python
customer_purchase = mydata.groupby('TransactionNo').agg(Quant = ('Quantity', 'sum'))
customer_purchase.sort_values(by = 'Quant', ascending = False).head(10).plot(kind= 'bar', stacked=True, figsize=(10,8),color='red', width = 0.5)

plt.xlabel('Transaction Number')
plt.ylabel('Number of Product')
plt.title('Chart Showing Number of Product bought per Transaction')
```
